# Project Scheduling
## Workers assignment

For this example, we are going to create a basic project scheduling model. We will take into account worker assignments and costs.

We consider the following instance:
- We have a project with a list of tasks that need to be performed 
- Each task has requirements to be fulfilled by the available workers
- Each worker has different skills and an expertise level needed to fulfill the requirements
- Each worker has a fixed and a variable cost for performing a requirement
- Tasks have precedence behavior; some tasks cannot start before another one starts or ends
- Parent tasks are defined, a parent tasks define the span of a group of tasks

## Using IBM Watson Studio to create a Decision Optimization Experiment

First of all, you need to log in to https://dataplatform.cloud.ibm.com/home 

As you log in, you are going to create your project. Go ahead and click on "Create a project".



Select the "Create an empty project" option.



Give a name to your project and select Cloud Object Storage. If you don't have one, create it and assign it to your project. A description can be added as well.



You are ready to create your model. Before creating the DO experiment, you are going to load the csv files with all the data needed to solve the problem example. Go to the Assets, to the right you will find the instructions to load your data. You can browse for files to upload or just drag and drop them in the box.



As loading finishes, data is now available in your project.



It's now time to create your DO Experiment. Click on "Add to project", then choose "Decision Optimization Experiment".



Now, you give a name to your DO Experiment and associate a Machine Learning service. 



For associating a Machine Learning service you have two options: create a new service or use an existing service. 



When a Machine Learning service is associated, you may now choose a deployment space for the experiment. If you don't have any deployment space, create it. Then click on "Create". 

If you are missing something for the experiment creation, the "Create" button will remain grey. As you finish all the needed steps to create the experiment, the "Create" button will turn blue.



Data must be prepared before creating our model. Select and import the data files already loaded to our proyect during the firts steps.



The data is now available for our DO Experiment.



Now, you need to crete the optimization model. Go to "Run model" and click on "Create"



Now you will find two options: Python or OPL. For this example you will choose OPL. 



You now may write your model.
